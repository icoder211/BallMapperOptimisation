```r
library(igraph)
library(scales)
library(networkD3)
library(stringr)
```

```r
library(BallMapper)
```

```r
x <- seq(from=0, to=6.24, by=0.1)
x
```

```r
points <- cbind(sin(x), cos(x))
plot(points)
values <- cos(x)
```

```r
points <- as.data.frame(points)
values <- as.data.frame(values)
epsilon <- 0.25
```

```r
l <- BallMapper(points, values, epsilon)
ColorIgraphPlot(l)
```
