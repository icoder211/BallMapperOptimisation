```python
library(BallMapper)
```

```python
x <- seq(from=0, to=6.24, by=0.1)
x
```

```python
points <- cbind(sin(x), cos(x))
plot(points)
values <- cos(x)
```

```python
points <- as.data.frame(points)
values <- as.data.frame(values)
epsilon <- 0.25
```

```python
l <- BallMapper(points, values, epsilon)
ColorIgraphPlot(l)
```

```python
df <- read.csv("boston_out.csv", header = FALSE)
head(df)
```

```python
values <- as.data.frame(df$V14)
head(values)
points <- subset(df, select = -V14)
head(points)
```

```python
epsilon <- 100
l <- BallMapper(points, values, epsilon)
ColorIgraphPlot(l)
```
